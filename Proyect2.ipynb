```python
#Import libraries
import pandas as pd
import pymongo
import csv
import json
import numpy as np
```

```python
#define de variable for each csv 
expo="resources/Exportacion.csv"
hibridos="resources/Vehi_hibrido_electricos.csv"
ventas="resources/Ventas.csv"
GEO="resources/GEO.json"
```

```python
#create the DF for each CSV
salesDF=pd.read_csv(ventas)
salesDF
```

```python
#DF for Hybrid and electric cars in Mexico (INEGI information)
HybrDF=pd.read_csv(hibridos,encoding='latin-1')
HybrDF
```

```python
#DF for car's exportations made in Mexico (INEGI information)

ExpoDF=pd.read_csv(expo)
ExpoDF
```

```python
#DF for countries coordinates to merge INEGI's information in D3
geos=pd.read_json(GEO)
mapacoords=geos.features.map(lambda x: (x["id"],x["geometry"]["coordinates"])).apply(pd.Series)
mapacoords
```

```python
#rename the columns 
coordsmap=mapacoords.rename(columns={0: "id", 1: "coordinates"})
coordsmap
```

```python
######## HYBRID'S SALES DATAFRAME BY YEAR #####################

HybByYear=HybrDF[["Año","Tema","EntidadFederativa","UnidadesVehicularesElectricas","UnidadesVehicularesHibridasPlugin1","UnidadesVehicularesHibridas2","Total"]]

SaleByType=HybByYear.groupby(["Año","EntidadFederativa"])
HybTotal=SaleByType["Total"].sum()
HybridByStateDF=pd.DataFrame({
    
    "Total_Sales":HybTotal
    
})
HybridByStateDF.head()

```

```python
HybridByStateDF.reset_index(inplace=True)
```

```python
HybridByStateDF.rename(columns={"Año":"Year","EntidadFederativa":"State","Total_Sales":"Sales"},inplace=True)
HybridByStateDF
```

```python
######## EXPORTATION DATAFRAME BY COUNTY AND YEAR #####################
ExpobyYear=ExpoDF[["Año","PaisDestino","Total"]]

ExpoByYear=ExpobyYear.groupby(["Año","PaisDestino"])
ExpoTotal=ExpoByYear["Total"].sum()
ExpCountriesdDF=pd.DataFrame({
    
    "Total_Sales":ExpoTotal
})
ExpCountriesdDF.head(5)
```

```python
ExpCountriesdDF.reset_index(inplace=True)
```

```python
ExpCountriesdDF.rename(columns={"Año":"Year","PaisDestino":"Country","Total_Sales":"Sales"},inplace=True)
ExpCountriesdDF
```

```python
#HybJson=HybridCars.to_json()
```

```python
######## SALES DATAFRAME BY BRAND AND YEAR #####################
TotalbyYear=salesDF[["Año","Marca","Total"]]

SaleByBrand=TotalbyYear.groupby(["Año","Marca"])
BrandTotal=SaleByBrand["Total"].sum()
ByBrandDF=pd.DataFrame({
    
    "Total_Sales":BrandTotal
})
ByBrandDF.head(5)
```

```python
ByBrandDF.reset_index(inplace=True)
```

```python
ByBrandDF.rename(columns={"Año":"Year","Marca":"Brand","Total_Sales":"Sales"},inplace=True)
ByBrandDF
```
